```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("simulation_data_5.csv")

plt.plot(data["Day"], data["Infected_deg"], label="Infected_deg")
plt.plot(data["Day"], data["Susceptible_deg"], label="Susceptible_deg")
plt.plot(data["Day"], data["Recovered_deg"], label="Recovered_deg")

plt.plot(data["Day"], data["Totall_deg"], label="Totall_deg")
plt.plot(data["Day"], data[" Vaccinated_deg"], label="Vaccinated_deg")

plt.xlabel("Day")
plt.ylabel("Population")
plt.title("Deggendorf Disease Simulation Results")
plt.legend()
plt.grid(True)
plt.show()
```
